1.Load the Data

```python
import pandas as pd
df = pd.read_csv('spotify.csv')  # Replace with your file name
```

2.Initial Clean-up

```python
df = df.dropna(subset=['popularity', 'danceability', 'energy', 'tempo', 'track_genre'])
df = df.drop_duplicates()
```

3.Visual Ideas

Distribution of Popularity

```python
df['popularity'].hist(bins=30)
```

Danceability vs Popularity Correlation

```python
df.plot.scatter(x='danceability', y='popularity')
```

Average Energy by Genre

```python
import matplotlib.pyplot as plt
genre_energy = df.groupby('track_genre')['energy'].mean().sort_values(ascending=False).head(10)
genre_energy.plot.bar()
plt.xlabel('Genre')
plt.ylabel('Average Energy')
```

Top Genres by Popularity

```python
genre_pop = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(10)
genre_pop.plot.bar()
```
